```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from requests import get
```

```python
headers = {"Accept-Language":"en-US,en;q=0.5"}
```

```python
url = "https://www.imdb.com/search/title/?groups=top_1000&ref_=adv_prv"
```

```python
import requests
```

```python
results = requests.get(url,headers=headers)
```

```python
soup = BeautifulSoup(results.text,"html.parser")
```

```python
#print(soup.prettify())
```

```python
titles = []
years = []
times = []
imbd_ratings = []
meta_scores = []
votes = []
us_gross = []
```

```python
movie_div = soup.find_all('div',class_ = 'lister-item mode-advanced')
```

```python
for container in movie_div:
    name = container.h3.a.text
    titles.append(name)
    year = container.h3.find('span',class_='lister-item-year').text
    years.append(year)
    runtime = container.find('span',class_='runtime').text if container.p.find('span',class_='runtime').text else "-"
    times.append(runtime)
    imbd = float(container.strong.text)
    imbd_ratings.append(imbd)
    meta_s = container.find('span',class_='metascore').text if container.find('span',class_='metascore') else "-"
    meta_scores.append(meta_s)
    nv = container.find_all('span',attrs={'name':'nv'})
    vote = nv[0].text
    votes.append(vote)
    grosses = nv[1].text if len(nv)>1 else '-'
    us_gross.append(grosses)
```

```python
df = pd.DataFrame({'movie':titles,'year':years,'rutime':runtime,
                  'imdb_rating':imbd_ratings,'meta_data':meta_scores,
                  'vote':votes,'us_gross_worth':us_gross})
```

```python
df['year'] = df['year'].str.extract('(\d+)').astype(int)
```

```python
df.dtypes
```

```python
df['rutime'] = df['rutime'].str.extract('(\d+)').astype(int)
```

```python
df.dtypes
```

```python
df['meta_data'] = df['meta_data'].astype(int)
```

```python
df.dtypes
```

```python
df['vote'] = df['vote'].str.replace(',','').astype(int)
```

```python
df.dtypes
```

```python
df['us_gross_worth'] = df['us_gross_worth'].map(lambda x: x.lstrip('$').rstrip('M'))
```

```python
df['us_gross_worth'] = pd.to_numeric(df['us_gross_worth'],errors='coerce')
```

```python
from time import sleep
from random import randint
```

```python
url = "https://www.imdb.com/search/title/?groups=top_1000&ref_=adv_prv"
```

```python
titles = []
years = []
time = []
imbd_ratings = []
metascores = []
votes = []
us_gross = []
```

```python
header = {'Accept-Language':'en-US,en;q=0.5'}
```

```python
pages = np.arange(1,1001,50)
```

```python
for page in pages:
    page = requests.get("https://www.imdb.com/search/title/?groups=top_1000&start=" + str(page)+ "&ref_=adv_nxt",headers=header)
    soup = BeautifulSoup(page.text,"html.parser")
    movie_div = soup.find_all('div',class_='lister-item mode-advanced')
    sleep(randint(2,10))
    for container in movie_div:
        name = container.h3.a.text
        titles.append(name)
        year = container.h3.find('span',class_='lister-item-year').text
        years.append(year)
        runtime = container.p.find('span',class_='runtime').text if container.p.find('span',class_='runtime').text else "-"
        time.append(runtime)
        imbd = float(container.strong.text)
        imbd_ratings.append(imbd)
        meta_s = container.find('span',class_='metascore').text if container.find('span',class_='metascore') else "-"
        metascores.append(meta_s)
        nv = container.find_all('span',attrs={'name':'nv'})
        vote = nv[0].text
        votes.append(vote)
        grosses = nv[1].text if len(nv)>1 else '-'
        us_gross.append(grosses)
```

```python
movies_df = pd.DataFrame({'titles':titles,'year':years,'runtime':time,
                         'imbd':imbd_ratings,'meta':metascores,'vote':votes,
                         'gross_amt':us_gross})
```
